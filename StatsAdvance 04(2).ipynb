```python
# 01
'''
The main difference between a t-test and a z-test is that a t-test is used when the sample size is small (usually less than 30) or when the population standard deviation is unknown, while a z-test is used when the sample size is large (typically greater than 30) and the population standard deviation is known.

Ex for a t-test: Comparing the average test scores of two groups of students (with less than 30 students in each group).

Ex for a z-test: Determining if there is a significant difference in the average height of men and women in a population (with large sample sizes for both genders).



'''
```

```python
# 02
'''
One-tailed test: It checks for a difference in one specific direction (e.g., better or worse).

Two-tailed test: It checks for a difference in any direction (e.g., just different, not specifying better or worse).
'''
```

```python
# 03
'''
Type 1 error: It occurs when we falsely reject a true null hypothesis. In other words, we conclude there's an effect when there isn't one.

Ex: Convicting an innocent person in a court trial.

Type 2 error: It happens when we fail to reject a false null hypothesis. In other words, we overlook an actual effect.

Ex: Acquitting a guilty person in a court trial.

'''
```

```python
# 04

'''
Bayes's theorem is a probability formula that helps update the probability of an event based on new evidence.
It is represented as: P(A|B) = [P(B|A) * P(A)] / P(B).

Ex: If we want to find the probability of a student being absent (A) given that it's raining (B),
and we know the probability of rain when a student is absent and the overall probability of a student being absent, 
we can use Bayes's theorem to update the probability.


'''
```

```python
# 08
'''
The margin of error in a confidence interval is the range within which the true population parameter is likely to lie. It represents the uncertainty in the estimate based on the sample data.

Sample size affects the margin of error inversely; a larger sample size reduces the margin of error, making the estimate more precise.

Ex: Surveying 200 customers instead of 50 yields a smaller margin of error for estimating satisfaction with a product.


'''
```

```python
'''
12:
Step 1: Setting up the hypothesis

    Null Hypothesis (H0): There is no significant difference in student performance between the two teaching methods.
    Alternative Hypothesis (H1): There is a significant difference in student performance between the two teaching methods.

Step 2:

    Sample A: Mean (average score) = 85, Standard Deviation = 6
    Sample B: Mean (average score) = 82, Standard Deviation = 5
    Significance level (α) = 0.01

Step 3: Calculate the t-statistic

t = (Mean of Sample A - Mean of Sample B) / sqrt[( (Standard Deviation of Sample A)^2 / Sample Size of A) + ((Standard Deviation of Sample B)^2 / Sample Size of B)]

let's assume both samples have 30 students each (nA = nB = 30):

t = (85 - 82) / sqrt[(6^2 / 30) + (5^2 / 30)]

t = 3 / sqrt[(36/30) + (25/30)]

t = 3 / sqrt[(1.2) + (0.83)]

t = 3 / sqrt(2.03)

t = 3 / 1.425

t = 2.105

Step 4: Find the critical t-value

As this is 2 tailed test with 0.01 significance level,the critical value will be:2.663

Step 5:Compare t-statistic and critical value

Since 2.105 < 2.663,we cannot reject the null hypothesis and we can say that there is no significant in student peformance bteween  2 teaching methods.









'''
```
